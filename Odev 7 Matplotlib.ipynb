```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
titan=pd.read_csv('titanic.csv', index_col=0, sep='\t')
titan
```

```python
titan1=titan.groupby(['Sex','Pclass']).count()
titan1
```

```python
plt.figure(figsize=(10,6))
a=list(titan1['Name'])
a[0:3] #[9,12,35]
a[3:] # [21,18,61]

plt.bar(np.arange(3)-0.2, a[3:], color='blue', width=0.3) #  buradaki ilk deger(height) np.arange(3) ile bir dizi seklinde
plt.bar(np.arange(3)+0.2, a[0:3], color='purple', width=0.3) #  tanimlanmistir. bu dizi ile uc grup sutunun merkezden saga ve
plt.title('Titanic Yolcular Cinsiyet-Yolcu Sinifi Iliskisi') # sola +-0.2 birim acilmasini sagladik. 
plt.xticks(np.arange(3), ['1.Sinif', '2.Sinif','3.Sinif'])

plt.xlabel('Yolcu Siniflari')
plt.ylabel('Yolcu Sayisi')

plt.legend(['Erkekler','Bayanlar'])
```
